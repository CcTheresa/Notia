

*   Construct Mapping Approach

This analysis transforms raw survey questions into higher-level behavioral constructs in order to model user engagement patterns in wellness and mental health mobile applications. Rather than analyzing individual survey items in isolation, conceptually related questions are grouped into theoretically meaningful constructs (e.g., emotional regulation, boredom-driven engagement, habit stability, self-improvement orientation).

Each construct represents an underlying psychological or behavioral driver inferred from multiple survey variables. Composite construct scores are computed by aggregating responses from relevant survey items, allowing complex engagement dynamics to be analyzed at a behavioral-system level rather than as fragmented variables.

This approach enables structured analysis of engagement mechanisms, supports cross-dataset comparison, and creates a bridge between behavioral data and product design decisions. It shifts the analysis from descriptive statistics to explanatory behavioral modeling.

```python
block_structure = {}

for block in q_blocks:
    block_cols = [col for col in df.columns if col.startswith(block)]
    block_structure[block] = block_cols

block_structure
```

```python
block_profile = {}

for block, cols in block_structure.items():
    block_profile[block] = {
        "num_columns": len(cols),
        "types": list(set(df[cols].dtypes.astype(str))),
        "sample_values": df[cols].head(3).to_dict()
    }

block_profile
```

```python
for block, meta in block_profile.items():
    print("\n", block)
    print("Columns:", meta["num_columns"])
    print("Types:", meta["types"])
    print("Sample:", list(meta["sample_values"].keys())[:3])
```
